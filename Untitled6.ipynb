```python
print(data['region'].unique())
print(data['parameter'].unique())
print(data['powertrain'].unique())
print(data['unit'].unique())
```

```python
data['region'] = data['region'].str.lower()
data['parameter'] = data['parameter'].str.lower()
data['powertrain'] = data['powertrain'].str.lower()
data['unit'] = data['unit'].str.lower()
```

```python
# Filter data for EV sales only
ev_sales_data = data[(data['parameter'] == 'ev sales') & (data['unit'] == 'vehicles')]

# Preview the filtered data
ev_sales_data.head()
```

```python
# Group by year and sum the sales to see the trend over time
yearly_sales = ev_sales_data.groupby('year')['value'].sum()

# Plot the yearly sales trend
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(x=yearly_sales.index, y=yearly_sales.values, marker='o')
plt.title('EV Sales Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Number of EV Sales')
plt.grid(True)
plt.show()
```

```python
# Group by region and sum the EV sales
region_sales = ev_sales_data.groupby('region')['value'].sum().sort_values(ascending=False)

# Plot the top 10 regions
plt.figure(figsize=(12, 6))
region_sales.head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Regions by EV Sales')
plt.xlabel('Region')
plt.ylabel('Number of EV Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the number of occurrences of each powertrain type in the EV sales data
powertrain_counts = ev_sales_data['powertrain'].value_counts()

# Plot the distribution of powertrain types
plt.figure(figsize=(8, 5))
sns.barplot(x=powertrain_counts.index, y=powertrain_counts.values, palette='viridis')
plt.title('Distribution of EV Powertrain Types')
plt.xlabel('Powertrain Type')
plt.ylabel('Count')
plt.show()
```

```python
# Group by region and sum the sales values
region_sales = ev_sales_data.groupby('region')['value'].sum().sort_values(ascending=False)

# Plot the top 10 regions by EV sales
plt.figure(figsize=(12, 8))
sns.barplot(x=region_sales.head(10), y=region_sales.head(10).index, palette='viridis')
plt.title('Top 10 Regions by EV Sales')
plt.xlabel('Number of EV Sales')
plt.ylabel('Region')
plt.grid(True)
plt.show()
```
